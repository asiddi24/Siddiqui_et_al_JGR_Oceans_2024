```python
data_dir = '/home/idies/workspace/Storage/asiddi24/persistent/SciServer/Chapter_1/Siddiqui_etal_JGR_2023/data/reviews_responses/'

ds_volintSPNA_200 = xr.open_dataset(data_dir+'ds_volintSPNA_200.nc')
ds_voltimeintSPNA_200 = xr.open_dataset(data_dir+'ds_voltimeintSPNA_200.nc')
ds_voltimeint_seasdtd_SPNA_200 = xr.open_dataset(data_dir+'ds_viltimeint_seasdtd_SPNA_200.nc')
```

```python
fig, axs = plt.subplots(nrows=3, ncols=1, sharex=True, figsize=(18,10))

axs = axs.flatten()


axs[0].grid(linestyle='-', linewidth='0.5', color='grey')
axs[0].plot(ds_volintSPNA_200.time, ds_volintSPNA_200.tnd*1e9, lw=3, color='black',label=r'S tendency')
axs[0].plot(ds_volintSPNA_200.time, ds_volintSPNA_200.adv*1e9, lw=3, color='grey',label=r'Advection')
axs[0].plot(ds_volintSPNA_200.time, ds_volintSPNA_200.dif*1e9, lw=3, color='orangered',label=r'Diffusion')
axs[0].plot(ds_volintSPNA_200.time, ds_volintSPNA_200.frc*1e9, lw=3, color='goldenrod',label=r'Forcing')
#axs[0].plot(ds_salt_anom_65_200.time, ds_salt_anom_65_200.__xarray_dataarray_variable__*1e-12, lw=3, color='blue',label=r'65$^{\circ}$N Advection')


#plt.ylim([-0.8,0.8])
axs[0].legend(loc='lower left', ncol=4, bbox_to_anchor=(0.1, 1.05),frameon=False,fontsize=20)
fig.subplots_adjust(left=0.1, right=0.78, bottom=0.1, top=0.9)

axs[0].set_ylabel('Vol. integrated\n  tendency \n[10$^{-9}$ g kg$^{-1}$ s$^{-1}$]', fontsize=22)

#axs[0].set_ylabel(r'[psu-kg] $\times$ 10$^{18}$', labelpad=20)
#axs[0].set_xlabel('Time [Yrs]')
axs[0].grid(True, alpha=0.4)
axs[0].tick_params(which='both',
               labelcolor='k', 
               labelsize=24, 
               direction='in',
               width=1,
               length=10,
               pad=15)
#axs[0].set_xticks([])

axs[0].xaxis.set_minor_locator(AutoMinorLocator())
axs[0].yaxis.set_minor_locator(AutoMinorLocator())
axs[0].tick_params(which='minor', direction='in', length=5, width=1, color='k')

axs[1].grid(linestyle='-', linewidth='0.5', color='grey')
axs[1].plot(ds_voltimeintSPNA_200.time, ds_voltimeintSPNA_200.tnd*1e-18, lw=3, color='black',label=r'$\mathcal{T}$')
axs[1].plot(ds_voltimeintSPNA_200.time, ds_voltimeintSPNA_200.adv*1e-18, lw=3, color='grey',label=r'$\mathcal{A}$')
axs[1].plot(ds_voltimeintSPNA_200.time, ds_voltimeintSPNA_200.dif*1e-18, lw=3, color='orangered',label=r'$\mathcal{M}$')
axs[1].plot(ds_voltimeintSPNA_200.time, ds_voltimeintSPNA_200.frc*1e-18, lw=3, color='goldenrod',label=r'$\mathcal{F}$')

# axs[1].set_ylim([-0.25,0.25])
#plt.yticks(np.arange(-1.2,1.8,0.6),fontsize=18)
#axs[1].legend(loc='upper left',bbox_to_anchor=(0.99, 0.8),frameon=False,fontsize=18)
fig.subplots_adjust(left=0.1, right=0.78, bottom=0.1, top=0.9)

axs[1].set_ylabel('Time integrated\n salt content\n [psu-kg] x 10$^{18}$',fontsize=22)
# axs[1].set_xlabel('Time [Yrs]')
#plt.ylim(-0.2,0.2)
axs[1].grid(True, alpha=0.4)
axs[1].tick_params(which='both',
               labelcolor='k', 
               labelsize=24, 
               direction='in',
               width=1,
               length=10,
               pad=15)


axs[1].xaxis.set_minor_locator(AutoMinorLocator())
axs[1].yaxis.set_minor_locator(AutoMinorLocator())
axs[1].tick_params(which='minor', direction='in', length=5, width=1, color='k')


axs[2].grid(linestyle='-', linewidth='0.5', color='grey')
axs[2].plot(ds_voltimeint_seasdtd_SPNA_200.time, ds_voltimeint_seasdtd_SPNA_200.tnd*1e-18, lw=3, color='black',label=r'$\mathcal{T}$')
axs[2].plot(ds_voltimeint_seasdtd_SPNA_200.time, ds_voltimeint_seasdtd_SPNA_200.adv*1e-18, lw=3, color='grey',label=r'$\mathcal{A}$')
axs[2].plot(ds_voltimeint_seasdtd_SPNA_200.time, ds_voltimeint_seasdtd_SPNA_200.dif*1e-18, lw=3, color='orangered',label=r'$\mathcal{M}$')
axs[2].plot(ds_voltimeint_seasdtd_SPNA_200.time, ds_voltimeint_seasdtd_SPNA_200.frc*1e-18, lw=3, color='goldenrod',label=r'$\mathcal{F}$')

# axs[2].set_ylim([-0.25,0.25])
#plt.yticks(np.arange(-1.2,1.8,0.6),fontsize=18)
#axs[1].legend(loc='upper left',bbox_to_anchor=(0.99, 0.8),frameon=False,fontsize=18)
fig.subplots_adjust(left=0.1, right=0.78, bottom=0.1, top=0.9)

axs[2].set_ylabel('Salt content\n anomaly\n [psu-kg] x 10$^{18}$',fontsize=22)
axs[2].set_xlabel('Time [Yrs]')
#plt.ylim(-0.2,0.2)
axs[2].grid(True, alpha=0.4)
axs[2].tick_params(which='both',
               labelcolor='k', 
               labelsize=24, 
               direction='in',
               width=1,
               length=10,
               pad=15)


axs[2].xaxis.set_minor_locator(AutoMinorLocator())
axs[2].yaxis.set_minor_locator(AutoMinorLocator())
axs[2].tick_params(which='minor', direction='in', length=5, width=1, color='k')


plt.figtext(0.16,0.13, r'\textbf{F$_1$}', fontsize=18)
plt.figtext(0.65,0.13, r'\textbf{F$_2$}', fontsize=18)

axs[0].axvspan(xmin='1992-01',xmax='1995-01',color='blue',alpha=0.1)
axs[0].axvspan(xmin='2000-01',xmax='2004-01',color='orange',alpha=0.1)
axs[0].axvspan(xmin='2011-12',xmax='2017-12',color='blue',alpha=0.1)

axs[1].axvspan(xmin='1992-01',xmax='1995-01',color='blue',alpha=0.1)
axs[1].axvspan(xmin='2000-01',xmax='2004-01',color='orange',alpha=0.1)
axs[1].axvspan(xmin='2011-12',xmax='2017-12',color='blue',alpha=0.1)

axs[2].axvspan(xmin='1992-01',xmax='1995-01',color='blue',alpha=0.1)
axs[2].axvspan(xmin='2000-01',xmax='2004-01',color='orange',alpha=0.1)
axs[2].axvspan(xmin='2011-12',xmax='2017-12',color='blue',alpha=0.1)


plt.savefig('figs/review_response_timeseries_detrending.png', format='png', bbox_inches='tight', facecolor='white', dpi=400)
```
